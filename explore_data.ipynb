```python
import matplotlib.pyplot as plt
import json
import os

from dataset import CaptionDataset
%matplotlib inline
```

```python
data_folder = './datasets/caption_data'
data_name = 'flickr8k_5_cap_per_img_5_min_word_freq'
```

```python
word_map_file = os.path.join(
    data_folder, 'WORDMAP_'+data_name+'.json')
with open(word_map_file, 'r') as f:
    word_map = json.load(f)
inver_word_map = {j:i for i, j in word_map.items()}
```

```python
caption_data = CaptionDataset(data_folder, data_name, split='TRAIN')
```

```python
img, caps, cap_lens = caption_data[0]
```
